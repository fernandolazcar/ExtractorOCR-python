Si queremos paginas en especifico de nuestro texto usaremos el codigo de **seleccion de las paginas a observar**
hay que pulsar control + enter, una vez cargo la primera parte nos pedira que ingresemos la pagina(s) que deseamos extraer  procedente a eso daremos enter e ingresaremos la pagina final.

````python
from pdf2image import convert_from_path
import pytesseract

# Ruta al archivo PDF que deseas convertir
pdf_path = '/content/capital expansive rate of growth.pdf'

# Convierte el PDF en una lista de imágenes
images = convert_from_path(pdf_path)

# Pregunta al usuario qué rango de páginas desean procesar
total_pages = len(images)
start_page = int(input(f"Ingresa el número de la primera página (1-{total_pages}): ")) - 1
end_page = int(input(f"Ingresa el número de la última página (1-{total_pages}): ")) - 1

if 0 <= start_page <= end_page < total_pages:
    # Procesa las imágenes de las páginas seleccionadas y extrae el texto
    for page_number in range(start_page, end_page + 1):
        selected_image = images[page_number]
        text = pytesseract.image_to_string(selected_image, lang='spa')
        print(f"Texto extraído de la página {page_number + 1}:\n{text}\n")
else:
    print("Rango de páginas inválido.")
````

Este codigo servira para **transformar el contenido en archivos Docx**
para insertar la ruta solo hay que introduccirla en esta linea:
Ruta al archivo PDF que deseas convertir
pdf_path = 'ruta del archivo pdf'

````python
from pdf2image import convert_from_path
import pytesseract
from docx import Document

# Ruta al archivo PDF que deseas convertir
pdf_path = '/content/prueba.pdf'

# Convierte el PDF en una lista de imágenes
images = convert_from_path(pdf_path)

# Crear un documento Word
doc = Document()

# Procesa cada imagen y extrae el texto
for i, image in enumerate(images):
    text = pytesseract.image_to_string(image, lang='spa')
    words = text.split()  # Dividir el texto en palabras
    clean_text = ' '.join(words)  # Unir las palabras con espacios
    print(f"Texto extraído de la página {i + 1}:\n{clean_text}\n")

    # Agregar el texto limpio al documento Word
    doc.add_paragraph(clean_text)

# Guardar el documento Word (.docx)
doc.save('archivo.docx')
````

Este codigo es lo mismo para el ejercicio anterior solo que en esta ocasion podremos seleccionar las paginas que son de nuestro interes

````python
from pdf2image import convert_from_path
import pytesseract
from docx import Document

# Ruta al archivo PDF que deseas convertir
pdf_path = '/content/prueba.pdf'

# Convierte el PDF en una lista de imágenes
images = convert_from_path(pdf_path)

# Crear un documento Word
doc = Document()

# Pregunta al usuario qué rango de páginas desean procesar
total_pages = len(images)
start_page = int(input(f"Ingresa el número de la primera página (1-{total_pages}): ")) - 1
end_page = int(input(f"Ingresa el número de la última página (1-{total_pages}): ")) - 1

if 0 <= start_page <= end_page < total_pages:
    # Procesa las imágenes de las páginas seleccionadas y extrae el texto
    for i in range(start_page, end_page + 1):
        image = images[i]
        text = pytesseract.image_to_string(image, lang='spa')
        words = text.split()  # Dividir el texto en palabras
        clean_text = ' '.join(words)  # Unir las palabras con espacios
        print(f"Texto extraído de la página {i + 1}:\n{clean_text}\n")

        # Agregar el texto limpio al documento Word
        doc.add_paragraph(clean_text)

    # Guardar el documento Word (.docx)
    doc.save('archivo.docx')
else:
    print("Rango de páginas inválido.")
````
